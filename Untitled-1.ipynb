# **Traffic Sign Detection (TSD)**

Meet our cutting-edge Traffic Sign Detector (TSD) system, revolutionizing road safety and traffic control. Using a Convolutional Neural Network (CNN) trained on a diverse dataset, it accurately spots and classifies signs in real-time and images. With adaptive learning and domain adaptation, it excels across varied environments, ensuring consistent, fast, and robust performance. The TSD system is a game-changer, contributing significantly to road safety and efficient traffic management.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import tensorboard
from PIL import Image
import keras
import os
from sklearn.model_selection import train_test_split 
from keras.utils import to_categorical
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout
from keras.preprocessing.image import ImageDataGenerator 
from keras.optimizers import Adam as AdamLegacy
from keras.callbacks import LearningRateScheduler

num_images_per_class = 2000
classes = 43

dataset_dir = "/Users/adityachaturvedi/Desktop/7th Sem/Minor Project"
output_dir = "/Users/adityachaturvedi/Desktop/7th Sem/Minor Project/CleanedData"

def clean_data(input_dir, output_dir, num_images_per_class):
    for i in range(classes):
        input_path = os.path.join(input_dir, 'train', str(i))
        output_path = os.path.join(output_dir, 'train', str(i))
        
        if not os.path.exists(output_path):
            os.makedirs(output_path)
        
        images = os.listdir(input_path)[:num_images_per_class]  
        for a in images:
            try:
                image = Image.open(os.path.join(input_path, a))
                image = image.resize((30, 30))
                image.save(os.path.join(output_path, a))
            except:
                print("Error cleaning image:", os.path.join(input_path, a))

clean_data(dataset_dir, output_dir, num_images_per_class)

def lr_schedule(epoch):
    initial_learning_rate = 0.001
    decay_rate = 0.95
    if epoch >= 10:
        return initial_learning_rate * decay_rate
    return initial_learning_rate

data = []
labels = []
classes = 43
cur_path = os.getcwd()

for i in range(classes):
    path = os.path.join(cur_path, 'train', str(i))
    images = os.listdir(path)[:num_images_per_class]  
    for a in images:
        try:
            image = Image.open(os.path.join(path, a))
            image = image.resize((30, 30))
            image = np.array(image)
            data.append(image)
            labels.append(i)
        except:
            print("Error loading image")


data = np.array(data)
labels = np.array(labels)
print(data.shape, labels.shape)

data = data / 255.0

X_t1, X_t2, y_t1, y_t2 = train_test_split(data, labels, test_size=0.2, random_state=42)
print(X_t1.shape, X_t2.shape, y_t1.shape, y_t2.shape)

plt.figure(figsize=(12, 12))
for i in range(43):
    class_indices = np.where(y_t1 == i)[0]
    plt.subplot(7, 7, i + 1)
    plt.imshow(X_t1[class_indices[0]])
    plt.title(f'Class {i}')
    plt.axis('off')
plt.show()


y_t1 = to_categorical(y_t1, 43)
y_t2 = to_categorical(y_t2, 43)
```

```python
from keras.models import Sequential

datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu', input_shape=X_t1.shape[1:]))
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(43, activation='softmax'))

optimizer = AdamLegacy(learning_rate=0.001)  
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

logdir="logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir)

eps = 50
batch_size = 32

lr_scheduler = LearningRateScheduler(lr_schedule)

history = model.fit(
    datagen.flow(X_t1, y_t1, batch_size=batch_size),
    steps_per_epoch=len(X_t1) // batch_size,
    epochs=eps,
    validation_data=(X_t2, y_t2),
    validation_steps=len(X_t2) // batch_size,
    callbacks=[lr_scheduler]
)

```

### **Model Summary**

*Data Augmentation for Robustness*

Extensive data augmentation techniques, including rotations, shifts, shearing, zooming, flips, and fill mode adjustments, are employed to enhance the model's resilience to variations in real-world traffic sign images.


#### *Convolutional Neural Network (CNN)*

The model architecture leverages a Convolutional Neural Network (CNN) to extract hierarchical features from input images:

Input Layer: Accepts 30x30 pixel RGB images.

Convolutional Layers: Two sets of convolutional layers with 32 filters (5x5) and ReLU activation functions.

Max-Pooling Layers: Spatial dimensions are reduced via 2x2 max-pooling layers.

Dropout Layers: Dropout layers (25% rate) mitigate overfitting.

Additional Convolutional Layers: Two subsequent sets of convolutional layers with 64 filters (3x3) and ReLU activations.

Flatten Layer: Feature maps are flattened for downstream fully connected layers.

Dense Layers: Two fully connected layers (256 units) with ReLU activations, followed by dropout layers (50% rate).

Output Layer: The final layer comprises 43 units with a softmax activation, corresponding to the 43 traffic sign classes.
Model Compilation


The model is compiled using the Adam optimizer with an initial learning rate of 0.001. Categorical cross-entropy serves as the loss function, while accuracy is the performance metric.


Training
Training spans 30 epochs, employing mini-batches of size 32. A learning rate scheduler adapts the learning rate during training. Augmented data generation diversifies the training data, facilitating better generalization.
This comprehensive architecture, in tandem with data augmentation and meticulous hyperparameter tuning, empowers the model to accurately classify traffic signs under diverse conditions. Further enhancements can be pursued through hyperparameter optimization, architectural refinements, and the expansion of the dataset to encompass more real-world scenarios.
